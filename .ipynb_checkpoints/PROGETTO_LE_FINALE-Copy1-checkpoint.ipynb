```python
# Функция для перевода строковых значений дней посещения приложения в числовые

def to_list_of_numbrs(list):
    if list != '':
        list = list.split(',')
        for numbr in range(len(list)):
            list[numbr] = int(list[numbr])
        return list
    else:
        return ''
```

Переводим столбцы в нужные данные, убирая NaN

Видим сильные выбросы пользователей по выручке в каждой из групп

Определим, какие пользователи попадают в эти выбросы. Для этого сначала определим 25% и 75% квартили для каждой из групп. Затем посчитаем межквартильный размах IQR. После этого можно будет вычислить границы "усов" из boxplot каждой группы. Всё, что находится за пределами "усов", и будет считаться выбросами

Так как выбросов снизу от коробок мы не наблюдаем ни в одной из групп, подсчитаем значение верхней границы "усов"

```python
test_q1 = users_test.query('total_revenue != 0').total_revenue.quantile(0.25)  # 8248.5
test_q2 =users_test.query('total_revenue != 0').total_revenue.quantile(0.75)  # 16428.75

IQR_t = test_q2 - test_q1

control1_q1 = users_control1.query('total_revenue != 0').total_revenue.quantile(0.25)  # 6132.75
control1_q2 = users_control1.query('total_revenue != 0').total_revenue.quantile(0.75)  # 12883.0

IQR_c1 = control1_q2 - control1_q1

control2_q1 = users_control2.query('total_revenue != 0').total_revenue.quantile(0.25)  # 6064.5
control2_q2 = users_control2.query('total_revenue != 0').total_revenue.quantile(0.75)  # 12727.0

IQR_c2 = control2_q2 - control2_q1
```

```python
upper_border_t = test_q2 + 1.5 * IQR_t  # 28699.125
upper_border_c1 = control1_q2 + 1.5 * IQR_c1  # 23008.375
upper_border_c2 = control2_q2 + 1.5 * IQR_c2  # 22720.75
```

Всё, что будет выше подсчитанных границ, будет являться выбросами. Найдём же их

```python
bros_t = users_test.query('total_revenue > @upper_border_t')
bros_c1 = users_control1.query('total_revenue > @upper_border_c1')
bros_c2 = users_control2.query('total_revenue > @upper_border_c2')
bros_t
```

Подсчитаем, какую долю от платящих пользователей занимают эти люди. Также подсчитаем, какая долю от общей выручки в группе приходится на этих пользователей

```python
users_t_no_bros = users_test.query('total_revenue <= @upper_border_t')
users_c1_no_bros = users_control1.query('total_revenue <= @upper_border_c1')
users_c2_no_bros = users_control2.query('total_revenue <= @upper_border_c2')
```

```python
bros_t.uid.nunique() / users_test.query('total_revenue != 0').uid.nunique()  # 8.1%
bros_c1.uid.nunique() / users_control1.query('total_revenue != 0').uid.nunique()  # 5.2%
bros_c2.uid.nunique() / users_control2.query('total_revenue != 0').uid.nunique()  # 4.8%
```

```python
bros_t.total_revenue.sum() / users_test.total_revenue.sum()  # 32.4% 
bros_c1.total_revenue.sum() / users_control1.total_revenue.sum()  # 34.9% 
bros_c2.total_revenue.sum() / users_control2.total_revenue.sum()  # 18.1% 
```

Во всех трёх группах выбросы приходятся на не менее 4% от от общего количества платящих пользователей и составляют не менее 18% выручки. Это очень большое значение, которое мы не можем просто проигнорировать.

Для того, чтобы оценить успешность А/В теста, необходимо выбрать нужные метрики








## Выберем метрики для сравнения.

### Целевые метрики:

- выручка с продажи не триал премиум-подписки
- средняя выручка на одного пользователя
- средняя выручка на одного платящего пользователя
- средняя выручка с покупки не триал премиума на одного пользователя
- средняя выручка с покупки не триал премиума на одного платящего пользователя

### Proxy-метрики:

- доля выручки от продажи не триал премиум-подписки
- доля платящих клиентов
- доля премиум-пользователей
- доля текущих премиум пользователей
- доля покупавших премиум (не триал) среди всех клиентов
- доля покупавших премиум (не триал) среди платящих клиентов

### guardrail-метрики:

- доля клиентов, отменивших премиум-подписку 

 До того как начать сравнивать тестовую группу с контрольными, нужно убедиться, что контрольные группы не различаются значимо по ключевым признакам и метрикам. 

Для определения значимости различий мы будем здесь и далее применять т-тесты различия средних, критерий хи-квадрат, а также бутстрап для различий в группах. В каждом из случаев будут проверяться гипотезы, соответствующие своему тесту

### Какие гипотезы для результатов сравнения мы выдвигаем:


#### Т-тест независимых выборок:

H0 - статистически значимых различий средних в исследуемых группах нет, p-value >= 0.05

H1 - в группах обнаружены статистически значимые различия средних, p-value < 0.05
 
 
  
#### Критерий хи-квадрат:

H0 - статичтически значимых различий в распределениях групп нет, p-value >= 0.05, распределения однородны

H1 - присутствуют статистически значимые различия в распределениях, p-value < 0.05, группы неоднородны


#### Бутстрап:

H0 - доверительный интервал пересекает 0, статистически значимой разницы в исследуемой метрике между группами нет

H1 - доверительный интервал не пересекает 0, между группами обнаружена статистически значимая разница в исследуемой метрике

```python
users_control1.uid.nunique()
# users_control2.uid.nunique()
```

Размеры выборок - 4340 и 4264, то есть примерно равные

Для сравнения долей платящих пользователей в контрольных группах используем критерий хи-квадрат

Создадим в каждой группе столбец, который будет равен 1, если пользователь платящий, и 0 - если нет. Также добавим столбец с обозначением группы

```python
users_test['is_payer'] = users_test['total_revenue'] > 0
users_test['is_payer'] = users_test.is_payer.astype(int)
users_test['group'] = 'test'

users_control1['is_payer'] = users_control1['total_revenue'] > 0
users_control1['is_payer'] = users_control1.is_payer.astype(int)
users_control1['group'] = 'control1'

users_control2['is_payer'] = users_control2['total_revenue'] > 0
users_control2['is_payer'] = users_control2.is_payer.astype(int)
users_control2['group'] = 'control2'
```

И теперь объединим две контрольные группы в одну таблицу для создания кросс-таблицы

```python
all_ctrls = pd.concat([users_control1, users_control2])
all_ctrls
```

Создадим кросс-таблицу по количеству пользователей (count), количеству платящих пользователей (sum) и доле платящих пользователей (mean)

```python
cross_payers = all_ctrls[all_ctrls.group.isin(['control1', 'control2'])].groupby('group')['is_payer']\
    .agg(['count', 'sum', 'mean'])
cross_payers
```

```python
st.chi2_contingency(cross_payers)
```

p > 0.05 - вероятно, различия между группами по этим показателям отсутствуют

Теперь, чтобы убедиться в этом окончательно, с помощью бутстрапа определим доверительный интервал у групп для разности долей платящих пользователей

```python
def diff_rate(x, y, axis=0):
    return np.mean(x, axis=axis) - np.mean(y, axis=axis)
```

```python
RPU = st.bootstrap(data=(users_control1.is_payer.values, users_control2.is_payer.values), statistic=diff_rate)
```

```python
RPU.confidence_interval
```

Интервал пересекает 0, а значит, он подтверждает наше предположение о том, что **значимых различий между контрольными группами нет**

Далее необходимо проверить, насколько корректно представлены страны в наших контрольных выборках. Создадим кросс-таблицу на основе количества пользователей из каждой страны, представим значения в каждом столбце как доли от общего количества в колонке

Теперь сделаем то же самое с поомощью pingouin

Для начала подготовим данные, сформировав сводную длинную таблицу из всех трёх групп

```python
marked_test = users_test[['uid', 'total_revenue']]
marked_test['group'] = 'test'

marked_control1 = users_control1[['uid', 'total_revenue']]
marked_control1['group'] = 'control1'

marked_control2 = users_control2[['uid', 'total_revenue']]
marked_control2['group'] = 'control2'

marked_control1
```

Здесь, конечно, уже ближе к нормальному распределению, но всё ещё далеко

```python
st.ttest_ind(avg_rev_of_payers_test.revenue, avg_rev_of_payers_control1.revenue, equal_var=True)
```

```python
st.ttest_ind(avg_rev_of_payers_test.revenue, avg_rev_of_payers_control2.revenue, equal_var=True)
```

Здесь т-тест показывает статистическую значимость различий для первой пары групп. Но, чтобы не делать поспешных выводов, применим поправку Геймса-Хоувелла

```python
pg.pairwise_gameshowell(data=avg_rev_of_payers_all_nbros, dv="revenue", between="group")
```

А вот тут уже видим для той же пары групп уровень значимости 0.07, что означает, что значимых различий, скорее всего, нет. Однако и тут мы не остановимся, а применим дополнительно бутстрап разницы средних

```python
boot_ARpremPPU_1 = st.bootstrap(data=(avg_rev_of_payers_test.revenue.values,
                                          avg_rev_of_payers_control1.revenue.values), 
                                  statistic=diff_rate)
boot_ARpremPPU_2 = st.bootstrap(data=(avg_rev_of_payers_test.revenue.values,
                                          avg_rev_of_payers_control2.revenue.values), 
                                  statistic=diff_rate)
```

```python
boot_ARpremPPU_1.confidence_interval
```

```python
boot_ARpremPPU_2.confidence_interval
```

Согласно результатам т-тестов и бутстрапа, между тестовой и контрольной группой различия **скорее всего есть**. Однако, тест по Геймсу-Хоувеллу дал отсутствие статистической значимости (p-value = 0.068). Значит, нельзя однозначно сказать, являются ли различия здесь статистически значимыми. Вероятнее всего, они находятся на пороге статистической значимости, и что-то однозначно утверждать здесь нельзя

Теперь найдём доли отмененных подписок среди тех, кто имел премиум-аккаунт, в каждой из групп

```python
cancel_rate_t = users_test.query('was_premium != 0 and is_premium == 0')\
    .uid.nunique() / users_test.query('was_premium != 0').uid.nunique()
cancel_rate_t
```

```python
cancel_rate_c1 = users_control1.query('was_premium != 0 and is_premium == 0')\
    .uid.nunique() / users_control1.query('was_premium != 0').uid.nunique()
cancel_rate_c1
```

```python
cancel_rate_c2 = users_control2.query('was_premium != 0 and is_premium == 0')\
    .uid.nunique() / users_control2.query('was_premium != 0').uid.nunique()
cancel_rate_c2
```

Здесь мы видим рост в тестовой группе относительно обеих контрольных групп. Нужно проверить, насколько значимо это увеличение

Сначала сформируем таблицы, состоящие только из платящих клиентов и сформируем там дополнительный столбец "cancel_prem", который равен 0, если пользователь не отменял премиум-подписку, и 1 - если отменял

```python
cancel_test = users_test.query('was_premium != 0')
cancel_test['cancel_prem'] = cancel_test.uid\
    .isin(cancel_test.query('was_premium == 1 and is_premium == 0')['uid'])
cancel_test['cancel_prem'] = cancel_test['cancel_prem'].astype(int)

cancel_control1 = users_control1.query('was_premium != 0')
cancel_control1['cancel_prem'] = cancel_control1.uid\
    .isin(cancel_control1.query('was_premium == 1 and is_premium == 0')['uid'])
cancel_control1['cancel_prem'] = cancel_control1['cancel_prem'].astype(int)

cancel_control2 = users_control2.query('was_premium != 0')
cancel_control2['cancel_prem'] = cancel_control2.uid\
    .isin(cancel_control2.query('was_premium == 1 and is_premium == 0')['uid'])
cancel_control2['cancel_prem'] = cancel_control2['cancel_prem'].astype(int)

cancel_control2.head()
```

Для использования критерия хи-квадрат, сформируем ещё 2 сводные длинные таблицы с пометкой группы данных

```python
all_cancels1 = pd.concat([cancel_test, cancel_control1])
all_cancels2 = pd.concat([cancel_test, cancel_control2])
```

Теперь используем критерий хи-квадрат для определения значимости изменений в тестовой группе

```python
cross_cancels1 = pd.crosstab(all_cancels1.cancel_prem, all_cancels1.group)
cross_cancels1
```

```python
st.chi2_contingency(cross_cancels1)
```

```python
cross_cancels2 = pd.crosstab(all_cancels2.cancel_prem, all_cancels2.group)
cross_cancels2
```

```python
st.chi2_contingency(cross_cancels2)
```

Видим p-value < 0.05 для второй пары групп. Вероятно, там изменения статистически значимы. Проверим это предположение при помощи бутстрапа

```python
boot_canc_prem_1 = st.bootstrap(data=(cancel_test.cancel_prem.values, 
                                      cancel_control1.cancel_prem.values), statistic=diff_rate)
boot_canc_prem_2 = st.bootstrap(data=(cancel_test.cancel_prem.values, 
                                      cancel_control2.cancel_prem.values), statistic=diff_rate)
```

```python
boot_canc_prem_1.confidence_interval
```

```python
boot_canc_prem_2.confidence_interval
```

Бутстрап подтвердил предположение о значимости изменений в тестовой группе относительно контрольной группы №2, так как доверительный интервал пересёк нулевое значение

```python
(cancel_rate_t - cancel_rate_c1) / cancel_rate_c1
```

```python
(cancel_rate_t - cancel_rate_c2) / cancel_rate_c2
```

   - 14.9% относительно контрольной группы №2. Изменение значительное

#### Далее пойдёт детальный разбор изменений в разрезе стран

Оставим только тех, кто купил премиум (не триал) подписку

```python
premium_no_trial_test = transactions_test.query('product_type == "premium_no_trial"')
premium_no_trial_control1 = transactions_control1.query('product_type == "premium_no_trial"')
premium_no_trial_control2 = transactions_control2.query('product_type == "premium_no_trial"')
premium_no_trial_test.head()
```

Найдём долю пользователей, купивших премиум от общего количества платящих клиентов по каждой стране. Исключаем страны, где менее 5 платящих клиентов для исключения нерепрезентативности.

```python
# Сначала группируем транзакции по странам
payers_test = transactions_test.groupby(['country', 'product_type']).agg({'uid': 'count'}).reset_index()

# Добавляем столбец с суммой транзакций внутри одной страны
payers_test['payers_sum'] = payers_test.groupby('country')['uid'].transform('sum')

# Узнаём долю покупаетелей премиум-аккаунта среди всех покупателей внутри одной страны
payers_test['user_rate'] = round(payers_test.uid / payers_test.payers_sum * 100)

# Оставляем только данные по покупателям премиум
payers_test = payers_test.query('product_type == "premium_no_trial" and payers_sum >= 5')
```

Теперь то же самое проделываем для обеих контрольных групп

```python
payers_control1 = transactions_control1.groupby(['country', 'product_type']).agg({'uid': 'count'}).reset_index()
payers_control1['payers_sum'] = payers_control1.groupby('country')['uid'].transform('sum')
payers_control1['user_rate'] = round(payers_control1.uid / payers_control1.payers_sum * 100)
payers_control1 = payers_control1.query('product_type == "premium_no_trial"')

payers_control2 = transactions_control2.groupby(['country', 'product_type']).agg({'uid': 'count'}).reset_index()
payers_control2['payers_sum'] = payers_control2.groupby('country')['uid'].transform('sum')
payers_control2['user_rate'] = round(payers_control2.uid / payers_control2.payers_sum * 100)
payers_control2 = payers_control2.query('product_type == "premium_no_trial"')

payers_control2
```

Соединяем данные от всех групп в одну таблицу, чтобы оставить только те страны, которые представлены во всех трёх группах

Столбцы пока что не переименовываем - сделаем это позже

```python
premium_no_trial_payers = pd.merge(payers_test, 
                                   pd.merge(payers_control1, payers_control2, 
                                            on=['country', 'product_type'], how='inner'), 
                                   on=['country', 'product_type'], how='inner')
premium_no_trial_payers
```

Разбиваем полученную таблицу на 3, оставляя только страну, количество пользователей и долю пользователей, чтобы пометить данные отдельно по каждой группе, а также переименовываем столбцы в единый вид.

```python
premium_no_trial_payers_test = premium_no_trial_payers[['country', 'uid', 'user_rate']]
premium_no_trial_payers_test['group'] = 'test'

premium_no_trial_payers_control1 = premium_no_trial_payers[['country', 'uid_x', 'user_rate_x']]
premium_no_trial_payers_control1['group'] = 'control_1'
premium_no_trial_payers_control1 = premium_no_trial_payers_control1\
    .rename(columns={'uid_x': 'uid', 'user_rate_x': 'user_rate'})

premium_no_trial_payers_control2 = premium_no_trial_payers[['country', 'uid_y', 'user_rate_y']]
premium_no_trial_payers_control2['group'] = 'control_2'
premium_no_trial_payers_control2 = premium_no_trial_payers_control2\
    .rename(columns={'uid_y': 'uid', 'user_rate_y': 'user_rate'})
```

И теперь добавляем данные из получившихся таблиц в одну длинную таблицу, где столбец 'group' будет показывать, к какой группе принадлежат данные. Это нужно, чтобы построить график далее

```python
countries_premium_no_trial_rate = pd.concat([premium_no_trial_payers_test, 
                                             premium_no_trial_payers_control1, 
                                             premium_no_trial_payers_control2], ignore_index=True)

countries_premium_no_trial_rate
```
